```python
# import required modules
import numpy as np
import pandas as pd
from faker import Faker

# assign Faker to a variable
fake = Faker()
# empty list to contain artists
artists = []
# artist counter to generate artist_id
artist_num = 0

#  department names and employe count in each as a dictionary
department_count = {
    'Modeling': 40,
    'Rigging':20,
    'Animation':60,
    'CFX':20,
    'FX': 60,
    'Lighting': 20,
    'Compositing': 80
}
# first unpack the department_count dict
for department, counts in department_count.items():
    # just a placeholder variable as we only need to access the head count for each dept to loop through
    for _ in range(counts):
        # for every iteration append the artists to the empty list we created in global space
        artists.append({
            # a simple f string to generate the id. artist_num +1 to not start the artist_id from 0
            "artist_id": f"A{artist_num+1:03}",
            # fake variable to generate new name in every iteration
            "artist_name": fake.name(),
            "department": department,
            "experience_level": np.random.randint(1, 15)
        })
        # increment the counter by 1 every iteration
        artist_num += 1

# assign the generated data to pandas DataFrame
artists_df = pd.DataFrame(artists)
```

```python
# write the data
artists_df.to_csv("../data/artists.csv", index=False)
```
