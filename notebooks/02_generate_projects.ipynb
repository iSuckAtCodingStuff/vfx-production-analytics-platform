```python

# GENEATE POJECTS

# Import required modules
import numpy as np
import pandas as pd
from faker import Faker
from datetime import timedelta

fake = Faker()
# create a lit with project names
project_names = [
    "Dragon Rider",
    "Bear in City",
    "Imaginary Bonds",
    "Galactic Guardians",
    "Sheep Quest",
    "Frozen History",
    "Teddy Bear",
    "Titan Rising",
    "Mystic Forest",
    "Solar Frontier"
]
# empty container for the projects data
projects = []

# studio names 
studio_names =[
            "Miller-Jones Entertainment",
            "Apex Studios",
            "Smith Media",
            "Positive Studios",
            "Quantum Entertainment",
            "Para-Mountain Entertainment",
            "Trixify Media House",
            "Blue Potato Entertainment",
            "Compass Media",
            "Story Store Studios"
        ]

# assign the random variable to studio names for assignment
random_studio = np.random.permutation(studio_names)
# loop using enumerate to get the index along with the value to help in creating project_id
for i, name in enumerate(project_names):
    project_type = np.random.choice([
            "Feature Film",
            "Streaming Series"
        ])
    # generate fake date range from 3 years till today
    start_date = fake.date_between(
        start_date="-3y",
        end_date="today"
    )
    if project_type == "Feature Film":
        budget = np.random.randint(80, 250)
        #duration of days the project will take/bidded days
        duration_days = np.random.randint(300, 900)
    else:
        budget = np.random.randint(15, 80)
        duration_days = np.random.randint(90, 500)
    end_date = start_date + timedelta(days=duration_days)
    
    # create condition for the project status 
    today = pd.Timestamp.today().date()
    if end_date < today:
        status = "Completed"
    else:
        status = "In Progress"

    projects.append({
        "project_id": f"P{i+1:03}",
        "project_name": name,
        "project_type": project_type,
        # random studio with the index of var i so that each studio is assigned once to a project
        "client": random_studio[i],
        "budget_million_usd": budget,
        "start_date": start_date,
        "end_date": end_date,
        "status": status
    })

projects_df = pd.DataFrame(projects)

projects_df
```

```python
# write to csv
projects_df.to_csv("..\data\projects.csv", index=False)
```
